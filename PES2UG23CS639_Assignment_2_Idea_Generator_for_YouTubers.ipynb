Import pipelines

```python
from transformers import pipeline, set_seed

```

Load GPT-2 Model

```python
generator = pipeline(
    "text-generation",
    model="gpt2"
)

set_seed(42)  # makes output reproducible
```

```python
def generate_titles(niche):
    prompt = f"""
List of viral YouTube video titles for the niche "{niche}":
1.
"""

    result = generator(
        prompt,
        max_length=120,
        num_return_sequences=1,
        temperature=0.9,
        top_k=50,
        top_p=0.95
    )

    return result[0]["generated_text"]
```

Test the Idea Generator

```python
output = generate_titles("Tech Review")
print(output)
```
